```python
import os
import pandas as pd
count=0
missing_trigerrs_files=[]
print("***************** start print ****************")
for root, sub_dirs, files in os.walk("C:/Users/hilla/Desktop/data/EEG Viewer"):
    print("root_directory_path", root)
    print("sub directories", sub_dirs)
    print("files", files)

    # run over the files in the current directory
    for file in files:
        full_file = os.path.abspath(os.path.join(root, file))
        print("full_file: ", full_file)
        if not file.endswith("Triggers.csv"):  # find triggers file
            continue
        print("file:" ,file)
       
        with open(full_file) as missing_triggers: #open the triggers file
            #lines = missing_triggers.readlines() #read every line in the file and save it as a list
            #print("lines:",lines)
            #print("type",type(lines))
            #print("type",type(lines[0]))
            triggers_file=pd.read_csv(full_file)
            print(triggers_file)
            print('length triggers file = ',len(triggers_file))
        if len(triggers_file) != 2 and len(triggers_file) != 12 and len(triggers_file) != 40 :
            print('in this file a trigger is missing',full_file)
            missing_trigerrs_files.append(full_file)
            count=count+1
            print('count = ',count)  
print('count = ',count)  
print('missing trigerrs files: ',missing_trigerrs_files)
```
